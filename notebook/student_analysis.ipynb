```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt



```

```python
df = pd.read_csv("../dat/students.csv")

df
```

```python
df["average"] = (df["math"] + df["english"] + df["science"]) / 3

df
```

```python
ranking = df.sort_values(by="average", ascending=False)

ranking
```

```python
plt.bar(df["name"], df["average"])

plt.title("Student Average Score")
plt.xlabel("Student")
plt.ylabel("Average Score")

plt.show()
```

```python
subjects_mean = df[["math", "english", "science"]].mean()

subjects_mean
```

```python
plt.hist(df["average"], bins=5)

plt.title("Distribution of Average Scores")
plt.xlabel("Average Score")
plt.ylabel("Students")

plt.show()
```
